```python
import pandas as pd
from ipywidgets import interact, Dropdown
import matplotlib.pyplot as plt

participants = {
    "ryan": ryan,
    "parthiv": parthiv,
}
```

```python

rows = []

for name, pdata in participants.items():
    for tp in ("pre", "post"):
        vas_list = pdata[tp]["vas"]
        vas_mean = sum(vas_list) / len(vas_list)

        rows.append({
            "participant": name,
            "timepoint": tp,
            "casf_mean": vas_mean,        # treating VAS as CAS-F here
            "sss": pdata[tp]["sss"],
            "pvt": pdata[tp]["pvt"],
        })

df = pd.DataFrame(rows)

# --- Group means across ALL participants for each timepoint (pre / post) ---

group_means = (
    df.groupby("timepoint")[["casf_mean", "sss", "pvt"]]
      .mean()
)

# --- Single graph with dropdown to choose measure ---

label_map = {
    "CAS-F": ("casf_mean", "CAS-F (group mean)", "Score"),
    "SSS": ("sss", "SSS (group mean)", "SSS"),
    "PVT": ("pvt", "PVT (group mean)", "Reaction time (ms)"),
}

def plot_measure(label):
    col, title, ylabel = label_map[label]

    plt.figure()
    group_means[col].plot(kind="bar")
    plt.title(title)
    plt.ylabel(ylabel)
    plt.xlabel("Timepoint")
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()

interact(
    plot_measure,
    label=Dropdown(
        options=list(label_map.keys()),
        value="CAS-F",
        description="Measure:",
    )
)
```
